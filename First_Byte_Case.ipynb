```python
# remove duplicate SKUs
items = df.drop_duplicates(subset = "SKU")
#print(items)
# determine sales_amount and operations_cost / divided by quantity
sales_per = items["Sales_Amount"] / items["Quantity"]
costs_per = items["Operational_Cost"] / items["Quantity"]
profit_per_item = sales_per - costs_per

# mean and median values
print("Mean Cost of Good:", costs_per.mean())
print ("Median Cost of Good:", costs_per.median())
print("Mean Sales From Good:", sales_per.mean())
print ("Median Sales From Good:", sales_per.median())
print("Mean Profit From Good:", profit_per_item.mean())
print ("Median Profit From Good:", profit_per_item.median())


# top 10 most profitable products
top_profits = items.groupby(profit_per_item).size().sort_values(ascending=False)
print("The items that generate the most profit are: ", top_profits.head(10))
## want to see SKU and SKU Category, not just profit
## graph

## least profitable items
lowest_profits = items.groupby(profit_per_item).size().sort_values(ascending=True)
print("The items that generate the least profit are: ", lowest_profits.head(10))
## graph

# repeat for costs and revenue (top and lowest 10)



## average profits graph
# graph seems wrong... -800 profit?
profit_fig = px.box(items, y = profit_per_item)
profit_fig.show()

## average revenue graph
revenue_fig = px.box(items, y = sales_per)
revenue_fig.show()
```
